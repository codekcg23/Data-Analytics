## SOM Algorithm

Input Vector (X) - connected with each neuron (unit) of a lattice/ map
Neighbourhood N(i) - Grow together

Connection (Synaptic connection) W - Conncections have weights that can be changed. (link between neuron and Input vector)
Neighbourhood function h_ij - Indicated how closely neuron i and k in output layer are connected to each other
### Competition
Winner (i) - Which is the closet vector who has most similar values to input vector. 

       - Here we find winner by a competition. So w.X need to be maximum. Because we find winner who is most Similar to W.

       - i(x) =  arg j min || X - w_j|| here =1,2,... m are units
### Colaboration
winner's neighbourhood neuron (j) - For each unit/neuron j in neighbourhood of the winning neuron i, we update the weights of j, w_j. So weight outside N(i) neighbourhood are not updated.

d_ij - lateral distance between winner i and neighbour unit j - euclidean distance

h_ij(d_ij) = exp(d_ij^2/ 2*sigma^)
### Learnng - weight update
w_j(n+1) = w_j(n) + adjustment 

adjustment = ita(n)* h_ij(x)(X-w_j(n))

ita - learing parameter

Concerge until no noticibale change

## Preprocessing

```python
X = df[['dim1','dim2']].values
print(X.shape)
X
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X = min_max_scaler.fit_transform(X)
X
```

```python
# select 2D grid with 5 by 5 nodes
lattice_size = (5,5)
lattice = np.random.random((*lattice_size,X.shape[1]))
learning_constant = 100
neighbourhood_constant = 100
sigma_init = 0.01
lr_rate_init = 0.01
n_iter = 1000
```

```python


def euclidian_distance(x,y):
    dist = np.sqrt(np.sum((x-y)**2))
    return dist

def competition(input_x):
    min_dist = 10000 # initialize min distance to mark neighbourhood
    for x in range(lattice.shape[0]):
        for y in range(lattice.shape[1]):
            unit = lattice[x,y].reshape(1,-1)   # neighbours
            input_x=input_x.reshape(1,-1)    # input x
            euc_dist= euclidian_distance(unit,input_x)  # find distance of winner and neigbour
            if euc_dist < min_dist:
                min_dist=euc_dist
                winner=lattice[x,y]
                winner_idx=np.array([x,y])
    return (winner, winner_idx)


def sigma_func(sigma_init,n,neighbourhood_constant):
    #n- iteration
    
    sigma = sigma_init * np.exp(-n/neighbourhood_constant)
    return sigma
    #distance = np.sqrt(np.sum((x-y)**2))



def learning_rate(lr_rate_init,n,learning_constant):
    
    # n - iter 
    lr_rate = lr_rate_init*np.exp(-n/learning_constant)
    return lr_rate

def collaboration(distance,sigma):
    return np.exp(-distance/ 2* math.pow(sigma,2))

def SOM (X,n_iter):
    for n in range(1,n_iter):
        N = X[np.random.randint(0,X.shape[0]),:]  # randomly select input vector
        winner,winner_ind = competition(N) 
        lr_rate = learning_rate(lr_rate_init,n,learning_constant)
        sigma = sigma_func(sigma_init,n,neighbourhood_constant)
        for x in range(lattice.shape[0]):
            for y in range(lattice.shape[1]):
                w=lattice[x,y].reshape(1,X.shape[1])
                w_dist=euclidian_distance(np.array([[x,y]]),winner_ind.reshape(1,2))
                print(w_dist)
                if w_dist<=sigma:  # with time we stop collobaration. Winner doesn't connitunig sharing all
                    influence=collaboration(w_dist,sigma)  # how much winer collabarete with it's neighbours
                    new_w=w+(lr_rate*influence*(N.reshape(1,-1)-w))
                    print('w',new_w)
                    lattice[x,y]=new_w.reshape(1,2)
                

```
